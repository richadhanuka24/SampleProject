```python
import pandas as pd
from datetime import date
df=pd.read_csv("car data.csv")
```

```python
df.head()
```

```python
df['new_Selling_Price']=df['Selling_Price']*100000
df.head()
```

```python
df['new_Present_Price']=df['Present_Price']*100000
df.head()
```

```python
from datetime import date
  
# creating the date object of today's date
todays_date = date.today()
print(todays_date)

df["Car_Age"]=todays_date.year-df['Year']
df.head()
```

```python
df.drop(["Seller_Type","Owner","Car_Name","Year","Selling_Price","Present_Price"],axis=1,inplace=True)
df.head()
```

```python
df=pd.get_dummies(df,drop_first=True)
df.head()
```

```python
X=df.drop(['new_Selling_Price'],axis=1)
X
```

```python
y=df['new_Selling_Price']
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=12)
```

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(X_train, y_train)
y_pred=svr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train, y_train)
y_pred=rf.predict(X_test)
r2_score(y_test,y_pred)
```

```python
import pickle
with open('car_model.pkl', 'wb') as model_file:
    pickle.dump(lr, model_file)
```
